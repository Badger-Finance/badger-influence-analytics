```python
from decimal import Decimal
from tabulate import tabulate

voting_data = {'total_cvx_sold_in_$': Decimal('106428.420'),
               'total_badger_sold_in_$': Decimal('40662.305'),
               'amount_cvx': Decimal('16893.400'),
               'amount_badger': Decimal('10453.035'),
               'badger_price_at_the_moment_of_sale_in_$': Decimal('3.890'),
               'cvx_price_at_the_moment_of_sale_in_$': Decimal('6.300'),
               '$/vlCVX for total votes': Decimal('0.060'),
               '$/vlCVX without badgerwbtc': Decimal('0.066')}
table_rows = [[
    voting_data['total_cvx_sold_in_$'], voting_data['total_badger_sold_in_$'], voting_data['amount_cvx'],
     voting_data['amount_badger'], voting_data['$/vlCVX for total votes'], voting_data['$/vlCVX without badgerwbtc'],]
]
table = tabulate(table_rows, tablefmt="github", headers=["Total amnt CVX $", "Total amnt BADGER $", "Amount CVX bought", "Amount BADGER bought","$/vlCVX total", "$/vlCVX w/o badgerwbtc votes"])
print(table)
```

```python
import matplotlib.pyplot as plt
from decimal import Decimal

plt.rcParams["figure.figsize"] = (10, 10)
bribes_data = {"frax": Decimal(228333.4003595493159628450234), "2pool-frax": Decimal(139675.8051593709082650805046),
               "dola": Decimal(617441.7348492212164141448451), "alusd": Decimal(563262.0933380010783766220852),
               "aleth": Decimal(565724.8043157838119237822106), "fpifrax": Decimal(112585.9844037608392463191238),
               "f-badgerwbtc": Decimal(235687.1553570462402634102000)}
labels = tuple(bribes_data.keys())
pools_bribes = list(bribes_data.values())
biggest_bribe_by_vote = max(pools_bribes)
biggest_bribe_ix = pools_bribes.index(biggest_bribe_by_vote)
explode = []
for bribe in pools_bribes:
    explode.append(0.1) if pools_bribes.index(bribe) == biggest_bribe_ix else explode.append(0)

fig1, ax1 = plt.subplots()
ax1.pie(pools_bribes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
